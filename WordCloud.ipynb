```python
from os import path
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
#read the txt file
text = open('reviews.txt').read()
```

```python
#create a wordcloud
wordcloud = WordCloud(min_font_size=1, max_font_size=120, background_color='white', ).generate(text)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
# this will remove the axis display along with the image.
plt.axis("off")
```
