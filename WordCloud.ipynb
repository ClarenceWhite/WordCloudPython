```python
from os import path
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
#read the txt file
contents = open('contents.txt').read()
headings = open('headings.txt').read()
```

```python
#create a wordcloud
wordcloud1 = WordCloud(min_font_size=1, max_font_size=120, background_color='white', ).generate(contents)
```

```python
plt.imshow(wordcloud1, interpolation='bilinear')
# this will remove the axis display along with the image.
plt.axis("off")
```

```python
#create a wordcloud
wordcloud2 = WordCloud(min_font_size=1, max_font_size=120, background_color='white', ).generate(headings)
```

```python
plt.imshow(wordcloud2, interpolation='bilinear')
# this will remove the axis display along with the image.
plt.axis("off")
```
